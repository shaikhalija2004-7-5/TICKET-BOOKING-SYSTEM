```python
print('----------------üöÜ INDIAN RAILWAY SYSTEM üöÜ---------------')
print('')
print('Rules of Indian Railways')
print("1. Always carry a valid ID proof.")
print("2. Travelling without a ticket is punishable.")
print("3. Below 5 years: No ticket required.")
print("9. Do not cross railway tracks; use foot-over bridges.")
print('-------------------------------------------------------------------')

## Inforamtion of passenger
def info():
    while True:
        PassengerName= input("Enter Your Name: ")
        print(f'Passenger name is {PassengerName.upper()}')
        print('')

        ##age info
        isMinor = False
        while True:
            Age = int(input('Enter Your Age: '))
            if Age<10 or Age >60:
                print('No tickets are requiredüëç') 
                print('No ticket are sanctioned for you! As per Indian Railway Rules! You are freely enjou trip')
                print('')
                isMinor = True
                break
            else:
                print(f'Age of Passenger is {Age} üëç')
                print('')
                break
        # print(isMinor)
        if(isMinor == True):
            continue
        
        ## adhar info
        while True:
            print('Only 12 digit number is required! ')
            Adhar_card_number = input('Enter Your Adhar Card Number')
            print(f'Adhar card number is {Adhar_card_number}')
            if len(Adhar_card_number) != 12 or not Adhar_card_number.isdigit():
                print('Adhar card number is not matched üëé! Type 12 digit number')
                print('')
                continue
            else :
                masked = Adhar_card_number[:4]+ 'xxxxxxxx'
                print(f'Adhar verifiedüëç {masked}')
                print('')
                break
        break

class Train:
    def __init__(self, tickets=10 , Fare = 350):
        self.A_tickets = tickets
        self.Exp = Fare
        

    def Availble_Seats(self):
        print(f'The available tickets of Duronto Express üöá is {self.A_tickets}')
        print('')

    def Booking(self):
        self.Tickets_book = int(input('How many Tickets do you want to book ? '))
        while True:
            if self.Tickets_book > self.A_tickets:
                print(f'Only! {self.A_tickets} seats are availbale')
                print('')
                continue
            else:
                self.A_tickets -= self.Tickets_book
                print(f'{self.Tickets_book} Tickets are booked üéüÔ∏è')
                print('')
                break

    def Ticket_fare(self):
        print(f'Per person ticket of Duronoto express is üöá {self.Exp} ')
        print(f'You booked {self.Tickets_book} Tickets üéüÔ∏è')
        print(f'Total charges of {self.Tickets_book} is {self.Tickets_book*self.Exp} üí∏')
        print('')

    def Exit(self):
        print(f'Thankyou for visiting Indian Railways üöÜ')
        print('')


RUN = Train()


print('------------ select option --------------')
print('1---------- For available Tickets')
print('2---------- For Book Tickets')
print('3---------- For Show Ticketss Fare')
print('4---------- For Exit')

while True:
    options = int(input("Enter your choice option"))
    if options ==1 :
        print(f'Your option No is {options} - For available Tickets')
        RUN.Availble_Seats()
        print('')
    elif options == 2:
        print(f'Your option No is {options} - For Book Tickets')
        RUN.Booking()
        print('')
    elif options == 3:
        print(f'Your option No is {options} - For Show Ticketss Fare')
        RUN.Ticket_fare()
        print('')
    else:
        print(f'Your option No is {options} - For Exit ')
        RUN.Exit()
        print('')
        break
```
